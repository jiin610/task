# LAB 8-1

#### datetime 모듈을 사용해 날짜/시간 출력

```python
import datetime

now = datetime.datetime.now()

print(f"오늘의 날짜 : {now.year}년 {now.month}월 {now.day}일")

if now.hour >= 12:
    ampm = "오후"
    hour = now.hour - 12 if now.hour > 12 else 12
else:
    ampm = "오전"
    hour = now.hour if now.hour > 0 else 12

print(f"현재시간 : {ampm} {hour}시 {now.minute}분 {now.second}초")
```

# LAB 8-2

#### 1. 크리스마스까지 남은 날짜 계산

```python
import datetime

now = datetime.datetime.now()
target = datetime.datetime(2027, 12, 25)
diff = target - now

hours = diff.seconds // 3600

print(f"오늘은 {now.year}년 {now.month}월 {now.day}일입니다.")
print(f"2027년 크리스마스 까지는 {diff.days}일 {hours}시간 남았습니다.")
```

#### 2. 1월 1일까지 남은 날짜 계산

```python
import datetime

now = datetime.datetime.now()
target = datetime.datetime(2036, 1, 1)
diff = target - now

hours = diff.seconds // 3600

print(f"오늘은 {now.year}년 {now.month}월 {now.day}일입니다.")
print(f"2036년 새해 까지는 {diff.days}일 {hours}시간 남았습니다.")
```

#### 3. 생일까지 남은 날짜 계산

```python
import datetime

now = datetime.datetime.now()
target = datetime.datetime(2026, 6, 10)
diff = target - now

hours = diff.seconds // 3600

print(f"오늘은 {now.year}년 {now.month}월 {now.day}일입니다.")
print(f"{target.year}년 생일까지는 {diff.days}일 {hours}시간 남았습니다.")
```

# LAB 8-3

#### 1. timedelta 클래스 사용해 오늘로부터 1000일 후 날짜 계산

```python
import datetime

now = datetime.datetime.now()
future = now + datetime.timedelta(days=1000)

print(f"오늘부터 1,000일 후의 날짜: {future.year}년 {future.month}월 {future.day}일")
```

#### 2. timedelta 클래스 사용해 100일 기념일 계산

```python
import datetime

date_input = input("처음으로 사귄 연도와 월, 일을 입력하시오 : ")
y, m, d = map(int, date_input.split())

start_date = datetime.datetime(y, m, d)
hundred_days = start_date + datetime.timedelta(days=99)

print(f"100일 기념일은 : {hundred_days.year}년 {hundred_days.month}월 {hundred_days.day}일입니다.")
```

#### 1. 4의 2승부터 10승까지 출력

```python
import math

for i in range(2, 11):
    print(f"4**{i:<2} = {math.pow(4, i)}")
```

#### 2. 0도에서 180도까지 라다인 값 출력

```python
import math

for i in range(0, 181, 10):
    print(f"{i} degree = {math.radians(i):.3f} radian")
```

#### 3. sin 0도에서 180도까지 출력

```python
import math

for i in range(0, 181, 10):
    print(f"sin({i:>3}) = {math.sin(math.radians(i)):.2f}")
```

# LAB 8-5

#### 1.randrange 함수를 사용해 0에서 100까지의 정수 중 5의 배수를 랜덤 세 개 추출

```python
import random

multiples_of_5 = []

for i in range(3):
    num = random.randrange(0, 101, 5)
    multiples_of_5.append(num)

print(f"0에서 100 이하의 정수 중에서 5의 배수 {multiples_of_5}")
```

#### 2. 1에서 10까지의 정수 중 랜덤으로 세 개 추출

```python
import random

random_integers = random.sample(range(1, 11), 3)

print(f"1에서 10 사이의 임의의 정수 : {random_integers}")
```

# LAB 8-6

#### 1. 로또번호 생성

```python
import random

print("이번 주의 추천 로또 번호")

for i in range(5):
    lotto = random.sample(range(1, 46), 6)
    lotto.sort()
    print(lotto)
```

#### 2. 로또번호와 보너스번호를 출력

```python
import random

print("이번 주의 당첨 로또 번호 + 보너스")

numbers = random.sample(range(1, 46), 7)

winning_numbers = numbers[:6]
winning_numbers.sort()

bonus_number = [numbers[6]]

print(f"{winning_numbers} + {bonus_number}")
```
